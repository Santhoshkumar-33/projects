```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
For demonstration purposes, we'll use a sample dataset:
```

```python
# Sample dataset
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.randint(1, 100, 50),
    'B': np.random.randint(1, 100, 50),
    'C': np.random.randn(50).cumsum(),
    'D': np.random.choice(['Category 1', 'Category 2', 'Category 3'], 50)
})
```

```python
# Line plot
plt.plot(data['A'], label='Line A')
plt.plot(data['B'], label='Line B')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
# Scatter plot
plt.scatter(data['A'], data['B'])
plt.xlabel('A')
plt.ylabel('B')
plt.title('Scatter Plot')
plt.show()
```

```python
# Bar Chart
counts = data['D'].value_counts()
plt.bar(counts.index, counts.values)
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Bar Chart')
plt.show()
```

```python
#Histogram
plt.hist(data['A'], bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#Pie Chart
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Boxplot
plt.boxplot([data['A'], data['B']])
plt.xticks([1, 2], ['A', 'B'])
plt.title('Box Plot')
plt.show()
```

```python
# Violin plot
plt.violinplot([data['A'], data['B']])
plt.xticks([1, 2], ['A', 'B'])
plt.title('Violin Plot')
plt.show()
```

```python
# Area plot
data[['A', 'B']].plot(kind='area', stacked=False)
plt.title('Area Plot')
plt.show()
```

```python
# Step plot
plt.step(data.index, data['A'], label='A')
plt.step(data.index, data['B'], label='B')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Step Plot')
plt.legend()
plt.show()
```

```python
# 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['A'], data['B'], data['C'])
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('C')
plt.title('3D Scatter Plot')
plt.show()
```

```python
# Customized line plot
plt.plot(data['A'], linestyle='--', color='r', marker='o')
plt.xlabel('Index')
plt.ylabel('A')
plt.title('Customized Line Plot')
plt.grid(True)
plt.show()
```
